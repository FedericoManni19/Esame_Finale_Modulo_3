```python
#Esercizio4
import pandas as pd
import seaborn as sns

# Carica il dataset se non già fatto
url = "https://raw.githubusercontent.com/owid/covid-19-data/refs/heads/master/public/data/owid-covid-data.csv"
df = pd.read_csv(url)

# Assicuriamoci che la data sia in formato datetime
df['date'] = pd.to_datetime(df['date'])

# Filtriamo per le nazioni richieste e il periodo
countries = ['Italy', 'Germany', 'France']

df_icu = df[
    (df['location'].isin(countries)) &
    (df['date'] >= '2022-05-01') &
    (df['date'] <= '2023-04-30')
]

# Tenere solo le colonne necessarie
df_icu = df_icu[['location', 'date', 'icu_patients']]

# Rimuoviamo i valori mancanti di ICU (giorni senza misurazioni)
df_icu = df_icu.dropna(subset=['icu_patients'])

# Controllo: quanti valori abbiamo per ogni nazione
print(df_icu.groupby('location')['icu_patients'].count())
```

```python
sns.set_style("whitegrid")

sns.boxplot(
    data=df_icu,
    x='location',
    y='icu_patients'
).set(
    title='Confronto pazienti in terapia intensiva (Mag 2022 - Apr 2023)',
    xlabel='Nazione',
    ylabel='Numero pazienti ICU'
)
```

```python
#Si osserva che Francia ha meno osservazioni rispetto a Italia e Germania, probabilmente per mancanza di dati giornalieri in alcune date. 
#Questo è normale nei dataset reali e non compromette il confronto generale tramite boxplot.
```

```python
#esercizio 5a
import pandas as pd
import seaborn as sns

# assicuriamoci che la data sia datetime
df['date'] = pd.to_datetime(df['date'])

countries = ['Italy', 'Germany', 'France', 'Spain']

df_hosp_2021 = df[
    (df['location'].isin(countries)) &
    (df['date'].dt.year == 2021)
]

# teniamo solo colonne utili
df_hosp_2021 = df_hosp_2021[['location', 'date', 'hosp_patients']]

# rimuoviamo i valori mancanti
df_hosp_2021 = df_hosp_2021.dropna(subset=['hosp_patients'])
```

```python
hosp_sum = df_hosp_2021.groupby('location')['hosp_patients'].sum().reset_index()

hosp_sum
```

```python
sns.set_style("whitegrid")

sns.barplot(
    data=hosp_sum,
    x='location',
    y='hosp_patients'
).set(
    title='Somma pazienti ospedalizzati nel 2021',
    xlabel='Nazione',
    ylabel='Somma hosp_patients'
)
```

```python
#esercizio5b
#Nel dataset sono presenti valori nulli per alcune nazioni, come la Germania, relativamente alla variabile hosp_patients nel 2021. In questo caso non è corretto sostituire i valori mancanti con zero, perché l’assenza del dato non implica assenza di pazienti ospedalizzati, ma solo mancanza di rilevazione. Pertanto, non è corretto effettuare una sostituzione che andrebbe a rendere non veritiera l’analisi ai dati effettivamente disponibili.
```
